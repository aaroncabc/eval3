# Comparación C vs Python

Cargar resultados de C y ejecutar un pipeline similar en Python para comparar etiquetas y UMAP.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import umap

c_labels = pd.read_csv('data/output/dbscan_labels.csv')
c_emb = pd.read_csv('data/output/umap_embedding.csv')
data = pd.read_csv('data/input/synthetic.csv')

print('DBSCAN labels C')
print(c_labels['label'].value_counts())

X = StandardScaler().fit_transform(data.values)
py_labels = DBSCAN(eps=0.75, min_samples=2).fit_predict(X)

print('DBSCAN labels Python')
print(pd.Series(py_labels).value_counts())

embedding = umap.UMAP(n_components=2, random_state=42).fit_transform(X)

fig, ax = plt.subplots(1, 2, figsize=(12, 5))
ax[0].scatter(c_emb['x'], c_emb['y'], c=c_labels['label'], cmap='tab10')
ax[0].set_title('UMAP C')
ax[1].scatter(embedding[:,0], embedding[:,1], c=py_labels, cmap='tab10')
ax[1].set_title('UMAP Python')
plt.show()
```
